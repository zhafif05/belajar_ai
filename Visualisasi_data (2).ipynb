```python

import matplotlib.pyplot as plt
import numpy as np


a = [60, 70, 80, 90, 100]
b = [55, 65, 60, 75, 95]

x = np.arange(len(a))
width = 0.35


plt.bar(x - width/2, b, width=width, color='r', label='Rakyat')


plt.bar(x + width/2, a, width=width, color='g', label='Koruptor')


plt.xlabel('Overs')
plt.ylabel('Runs Scored')
plt.title('Scoring Chart Rakyat & Koruptor')
plt.xticks(x, ['1-10', '11-20', '21-30', '31-40', '41-50'])  # Label interval over


plt.legend()

plt.show()
```

```python

import matplotlib.pyplot as plt


slices = [32, 18, 15, 18, 5]
departments = ['Dpr', 'Bumn', 'Bupati', 'Gubernur', 'Bendahara Umum Negara (BUN)']
colors = ['orange', 'gold', 'green', 'red', 'blue']
explode = [0.1, 0, 0, 0, 0]


plt.pie(slices,
        labels=departments,
        colors=colors,
        startangle=90,
        explode=explode,
        shadow=True,
        autopct='%1.1f%%')

plt.title('Koruptor di Indonesia')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = range(4)
y = [20.5, 30, 28, 40]
labels = ['Q1', 'Q2', 'Q3', 'Q4']

plt.bar(x, y, width=0.25)
plt.xticks([0, 1, 2, 3], labels)
plt.title('Bar Chart with Custom X-Ticks')
plt.xlabel('Quarters')
plt.ylabel('Sales (in thousands)')
plt.show()
```
