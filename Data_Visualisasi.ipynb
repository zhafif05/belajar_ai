# Case 1

Home store mana yang banyak mendapatkan customer?

```python
# hitung jumlah customer tiap home store
home_store_counts = df_cust['home_store'].value_counts()

print(home_store_counts)

# kalau mau lebih rapi jadi DataFrame
home_store_df = df_cust.groupby('home_store')['customer_id'].count().reset_index()
home_store_df.columns = ['HomeStore', 'JumlahCustomer']
home_store_df = home_store_df.sort_values(by='JumlahCustomer', ascending=False)

print(home_store_df)
```

Buatkan Visualisasi nya menggunakan bar plot atau chart yang lain. Kemudian jelaskan insight apa yang bisa di dapatkan!



```python
sns.barplot(x="HomeStore", y="JumlahCustomer", data=home_store_df)
plt.title("Jumlah Customer per Home Store")
plt.show()
```

# Case 2

Distribusi umur dari masing-masing home store menggunakan boxplot  

```python
from datetime import datetime

# Hitung umur berdasarkan tahun sekarang
tahun_sekarang = datetime.now().year
df_cust['age'] = tahun_sekarang - df_cust['birth_year']

# Cek data
print(df_cust[['home_store','birth_year','age']].head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.boxplot(x="home_store", y="age", data=df_cust, palette="Set2")

plt.title("Distribusi Umur Customer per Home Store")
plt.xlabel("Home Store")
plt.ylabel("Umur Customer")
plt.show()
```
