#### Data Preprocesing

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Lakukan Label Encoding untuk kolom kategorikal
df['workclass'] = label_encoder.fit_transform(df['workclass'])
df['education'] = label_encoder.fit_transform(df['education'])
df['marital.status'] = label_encoder.fit_transform(df['marital.status'])
df['occupation'] = label_encoder.fit_transform(df['occupation'])
df['relationship'] = label_encoder.fit_transform(df['relationship'])
df['race'] = label_encoder.fit_transform(df['race'])
df['sex'] = label_encoder.fit_transform(df['sex'])
df['native.country'] = label_encoder.fit_transform(df['native.country'])
df['income'] = label_encoder.fit_transform(df['income'])

# Menampilkan daftar label unik terakhir yang di-encode
label_encoder.classes_

# Melihat hasil encoding dari satu kolom (contohnya 'workclass')
df['workclass']
```

##### Missing Value

```python
# Isi missing value di kolom numerik dengan mean
df['age'] = df['age'].fillna(df['age'].mean())
df['fnlwgt'] = df['fnlwgt'].fillna(df['fnlwgt'].mean())
df['education.num'] = df['education.num'].fillna(df['education.num'].mean())
df['capital.gain'] = df['capital.gain'].fillna(df['capital.gain'].mean())
df['capital.loss'] = df['capital.loss'].fillna(df['capital.loss'].mean())
df['hours.per.week'] = df['hours.per.week'].fillna(df['hours.per.week'].mean())

# Cek apakah masih ada missing value
print(df.isnull().sum())
```

##### Outlier

```python
# Daftar kolom numerik yang mungkin memiliki outlier
features_with_outliers = ['age', 'fnlwgt', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']

# Proses Winsorizing (mengganti outlier dengan batas bawah/atas)
for feature in features_with_outliers:
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    IQR = q3 - q1
    lower_limit = q1 - (IQR * 1.5)
    upper_limit = q3 + (IQR * 1.5)

    # Ganti nilai outlier dengan batas bawah atau atas
    df.loc[df[feature] < lower_limit, feature] = lower_limit
    df.loc[df[feature] > upper_limit, feature] = upper_limit

print("Outlier sudah ditangani dengan metode Winsorizing.")
```

```python
# Ambil semua kolom numerik
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Tentukan jumlah kolom & baris subplot (misalnya 3 kolom per baris)
n_cols = 3
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols  # ceiling division

plt.figure(figsize=(15, 5*n_rows))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.boxplot(df[col], vert=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```
