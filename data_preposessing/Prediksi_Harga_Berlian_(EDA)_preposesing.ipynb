#### Data Preprocesing

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['warna'] = label_encoder.fit_transform(df['warna'])
df['potongan'] = label_encoder.fit_transform(df['potongan'])
df['kejernihan'] = label_encoder.fit_transform(df['kejernihan'])
label_encoder.classes_
df['warna']
```

##### Missing Value

```python
df['warna'] = df['warna'].fillna(df['warna'].mean())
df['warna'].isnull()

df['x'] = df['x'].fillna(df['x'].mean())
df['y'] = df['y'].fillna(df['y'].mean())
df['z'] = df['z'].fillna(df['z'].mean())

df.isnull().sum()
```

##### Outlier

```python
features_with_outliers = ['carat', 'kedalaman', 'lebar', 'x', 'y', 'z', 'harga']

for feature in features_with_outliers:
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    IQR = q3 - q1
    lower_limit = q1 - (IQR * 1.5)
    upper_limit = q3 + (IQR * 1.5)

    # Winsorizing: mengganti outlier dengan batas bawah/atas
    df.loc[df[feature] < lower_limit, feature] = lower_limit
    df.loc[df[feature] > upper_limit, feature] = upper_limit
```

```python
# Ambil semua kolom numerik
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Tentukan jumlah kolom & baris subplot (misalnya 3 kolom per baris)
n_cols = 3
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols  # ceiling division

plt.figure(figsize=(15, 5*n_rows))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.boxplot(df[col], vert=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```
